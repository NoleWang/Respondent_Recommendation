## Retreive users from Gitter

```python
time_to_sleep = 10*60
```

```python
def user_retreive(users_link,skip_num,roomId):
    usernames = []
    users = []
    #count = 0
    while True:
        api_link = requests.get(users_link)
        users_data = api_link.json()
        if len(users_data) == 0:
            break      
        elif len(users_data) == 1 and "name" not in json.dumps(users_data):
            if "Too Many Requests" in users_data["error"]:
                #time_to_sleep = (60 - datetime.now().minute + 1)*60
                print("Reach limit! Sleep ",time_to_sleep, " seconds!")
                time.sleep(time_to_sleep)
                users_link = "https://api.gitter.im/v1/rooms/"+roomId+"/users?access_token="+gitter_token+"&limit=100&skip="+str(skip_num)
                continue
            break    
            
        #print(len(users_data))
        for num in range(len(users_data)):
            #print(count)
            #print(users_data[num])
            
            dict_user = {}
            dict_user["id"] = users_data[num]["id"]
            dict_user["username"] = users_data[num]["username"]
            usernames.append(users_data[num]["username"])
            dict_user["displayName"] = users_data[num]["displayName"]
        
            users.append(dict_user)
            #count += 1
    
        skip_num += 100
        users_link = "https://api.gitter.im/v1/rooms/"+roomId+"/users?access_token="+gitter_token+"&limit=100&skip="+str(skip_num)
    
    print("Total users: ",len(users))
    return users,usernames
```

## Retreieve messages from Gitter

```python
def msg_retreive(msg_link,beforeId,roomId):
    chatlogs = []
    #count = 0
    while True:
        api_link = requests.get(msg_link)
        chat_data = api_link.json()
        if len(chat_data) == 0:
            break      
        elif len(chat_data) == 1 and "text" not in json.dumps(chat_data):
            if "Too Many Requests" in chat_data["error"]:
                #time_to_sleep = (60 - datetime.now().minute + 1)*60
                print("Reach limit! Sleep ",time_to_sleep, " seconds!")
                time.sleep(time_to_sleep)
                msg_link = "https://api.gitter.im/v1/rooms/"+roomId+"/chatMessages?access_token="+gitter_token+"&limit=100&beforeId="+beforeId
                continue
            break    
        else:
            beforeId = chat_data[0]["id"]
        
        chatlogs.extend([chat_data[num] for num in range(len(chat_data))])
        #print(len(chatlogs))
        
        msg_link = "https://api.gitter.im/v1/rooms/"+roomId+"/chatMessages?access_token="+gitter_token+"&limit=100&beforeId="+beforeId
    
    print("Total messages: ",len(chatlogs))
    return chatlogs
```

## Retreive user profile from GitHub

```python
with open(GitHub_folder+'git_account.txt', 'r') as fp:
    credentials = fp.readlines()
```

```python
account = credentials[0]
token = credentials[1]
```

```python
def git_profile(usernames, account,token):
    time_to_sleep = 8*60
    users_not_found = []
    
    for username in usernames:
        username = username.strip()
        if os.path.exists(GitHub_folder+channel2+"/"+username+'.json') or username in set(users_not_found):
            continue
    
        #rate limit default = 30 per page, retreive data from all pages
        page = 1
        link = "https://api.github.com/users/"+username+"/repos?per_page=100&page="+str(page)
        repos = []
        count = 1
        while True:
            api_link = requests.get(link, auth=(account,token))
            api_data = api_link.json()
            
            if api_link.status_code == 404 or len(api_data) == 0:
                break
            
            if "name" not in json.dumps(api_data):
                if "API rate limit exceeded" in api_data["message"]:
                    #time_to_sleep = (60 - datetime.now().minute + 1)*60
                    print("Reach limit in profile! Sleep ",time_to_sleep, " seconds!")
                    time.sleep(time_to_sleep)
                    link = "https://api.github.com/users/"+username+"/repos?per_page=100&page="+str(page)
                    continue
                elif "Repository access blocked" in api_data["message"]:
                    print("Repository access blocked!")
                    break
            
                
            
            #print(username, page, len(api_data))
            for num in range(len(api_data)):
                dict_user = {}
                flag = True
                repo_name = api_data[num]["name"]
                #print(count, ": ",repo_name)
                #readme_link = "https://api.github.com/repos/"+username+"/"+repo_name+"/readme"
                lan_link = "https://api.github.com/repos/"+username+"/"+repo_name+"/languages"
                #data = requests.get(lan_link, auth=(account,token))
                
                while flag:
                    try:
                        data = requests.get(lan_link, auth=(account,token))
                        flag = False
                    except ConnectionError as e:
                        print("Connect Error!")
                        print("Sleep 100 seconds!")
                        time.sleep(100)
                
                if data.status_code != 200:
                    print("!!!!!!!",data.status_code)
                lan_data = data.json()
                
                while "message" in json.dumps(lan_data) and "API rate limit exceeded" in lan_data["message"]:
                    print("Reach limit in lan! Sleep ",time_to_sleep, " seconds!")
                    time.sleep(time_to_sleep)
                    flag = True
                    
                    while flag:
                        try:
                            data = requests.get(lan_link, auth=(account,token))
                            flag = False
                        except ConnectionError as e:
                            print("Connect Error!")
                            print("Sleep 100 seconds!")
                            time.sleep(100)
                            
                    lan_data = data.json()
                
        
                dict_user["name"] = repo_name
                dict_user["url"] = api_data[num]["html_url"]
                dict_user["description"] = api_data[num]["description"]
                dict_user["language"] = list(lan_data.keys())
                dict_user["forks_count"] = api_data[num]["forks_count"]
                dict_user["stargazers_count"] = api_data[num]["stargazers_count"]
                dict_user["watchers_count"] = api_data[num]["watchers_count"]
                dict_user["size"] = api_data[num]["size"]
                dict_user["topics"] = api_data[num]["topics"]
        
                repos.append(dict_user)
                count += 1
            page += 1
            link = "https://api.github.com/users/"+username+"/repos?per_page=100&page="+str(page)
        
    
        print(username,": ",len(repos)," repos")
        if len(repos) == 0:
            users_not_found.append(username)
        else:
            with open(GitHub_folder+channel2+"/"+username+'.json', 'w') as outfile:
                json.dump(repos, outfile)
    return users_not_found
```

## User metadata

```python
import pandas as pd
import numpy as np
```

```python
def create_profile(path):
    names= []
    repos=[]
    languages = []
    topics=[]
    
    for username in os.listdir(path):
        names.append(username[:-5])
        
        with open(path+username, 'r') as f:
            data = json.load(f)
            lang = set()
            topic = set()
            repos.append(len(data))
        
            for num in range(len(data)):
                lang.update(lang,data[num]["language"])
                topic.update(topic,data[num]["topics"])
            
            languages.append(list(lang))
            topics.append(list(topic))
    
    df = pd.DataFrame()
    df["User"] = names
    df["Repo_num"] = repos
    df["Repo_languages"] = languages
    df["Repo_topics"] = topics
    #print(df.head(10))
    
    return df
```
